```python
#importing all required libraries
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#loading and preprocessing data
df = pd.read_csv("Datasets/Regression/Walmart_Sales.csv")
df = df.drop("Date", axis=1)
df = df.drop("Store", axis=1)
df.head()
```

```python
#linear regression from scratch
#X and y including their train test splits have already been made during the preprocessing step
def createWeightsAndBias(X):
    W = np.zeros(X.shape[1])
    B = 1
    return W, B


def updateWeights(X, y, W, B, alpha = 0.001):
    indices = [i for i in range(X.shape[0])]
    np.random.shuffle(indices)
    cost = 0
    for i in indices:
        yhat = np.dot(X[i], W) + B
        cost += ((yhat-y[i])**2)/2
        for j in range(X.shape[1]):
            W[j] -= alpha*((yhat - y[i])*X[i][j])
        B -= alpha*((yhat - y[i]))
    return W, B, cost

def train(X, y, epochs = 10):
    W, B = createWeightsAndBias(X)
    lossDuringTraining = []
    epoch = [0]
    lossDuringTraining.append(np.sum(((np.dot(X, W)+B-y)**2)/2))
    for j in range(epochs):
        W, B, temp = updateWeights(X, y, W, B)
        lossDuringTraining.append(temp)
        epoch.append(j)
    return W, B, lossDuringTraining, epoch
```
